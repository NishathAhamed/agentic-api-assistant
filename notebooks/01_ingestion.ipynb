```python
import sys
sys.path.append("../")

from document_extract.extractor import extract_text

def show_result(path: str):
    doc = extract_text(path)
    print("="*80)
    print("FILE:", path)
    print("filetype:", doc.filetype)
    print("pages:", doc.pages)
    print("warnings:", doc.warnings)
    print("\nFirst 500 chars:\n")
    print(doc.text[:500])
    print("="*80)
    return doc
```

```python
doc_docx = show_result("../assets/Quick Wins – Email Change Api Documentation.docx")
assert doc_docx.filetype == "docx"
assert len(doc_docx.text.strip()) > 20
assert doc_docx.warnings == []
```

```python
doc_pdf = show_result("../assets/Genie Service Request API Doc V2 1 (1) (1).pdf")
assert doc_pdf.filetype == "pdf"
assert doc_pdf.pages is not None and doc_pdf.pages > 0
# pdf text might be empty if it is scanned; so check rules:
if len(doc_pdf.text.strip()) == 0:
    assert "needs_ocr" in doc_pdf.warnings
else:
    assert doc_pdf.warnings == []
```
